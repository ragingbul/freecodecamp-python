```python
class LinkedList:

    # ---------------- NODE CLASS ----------------
    # Each Node represents one element in the linked list.
    class Node:

        def __init__(self, element):
            # Store the data/value of the node.
            self.element = element

            # Pointer to the next node.
            # Initially, the node doesn't point anywhere.
            self.next = None

    # ---------------- LINKED LIST CONSTRUCTOR ----------------
    def __init__(self):

        # Keeps track of how many nodes are in the list.
        self.length = 0

        # Head points to the first node.
        # Since the list is empty initially, head is None.
        self.head = None

    # ---------------- CHECK IF LIST IS EMPTY ----------------
    def is_empty(self):

        # Returns True if there are no nodes.
        # Returns False otherwise.
        return self.length == 0

    # ---------------- ADD A NODE ----------------
    def add(self, element):

        # Create a new node containing the given element.
        node = self.Node(element)

        # If the list is empty,
        # the new node becomes the head.
        if self.is_empty():
            self.head = node

        # Otherwise, traverse to the last node.
        else:

            # Start from the head.
            current_node = self.head

            # Continue moving until the last node.
            while current_node.next is not None:
                current_node = current_node.next

            # Connect the last node to the new node.
            current_node.next = node

        # Increase the size of the linked list.
        self.length += 1

    # ---------------- REMOVE A NODE ----------------
    def remove(self, element):

        # previous_node keeps track of the node before current_node.
        previous_node = None

        # Start searching from the head.
        current_node = self.head

        # Traverse the list until:
        # 1. We reach the end
        # OR
        # 2. We find the element.
        while current_node is not None and current_node.element != element:

            # Remember the current node.
            previous_node = current_node

            # Move to the next node.
            current_node = current_node.next

        # If we reached the end of the list,
        # the element wasn't found.
        if current_node is None:
            return

        # If previous_node exists,
        # the node to remove is NOT the head.
        elif previous_node is not None:

            # Skip over current_node.
            #
            # Example:
            # 1 -> 2 -> 3
            #
            # Removing 2:
            #
            # previous_node.next = current_node.next
            #
            # becomes
            #
            # 1 -------> 3
            previous_node.next = current_node.next

        # Otherwise,
        # we're removing the head node.
        else:

            # Move the head to the second node.
            self.head = current_node.next

        # One node has been removed,
        # so decrease the length.
        self.length -= 1


# ---------------- TESTING ----------------

# Create an empty linked list.
my_list = LinkedList()

# Should print True.
print(my_list.is_empty())

# Add two nodes.
my_list.add(1)
my_list.add(2)

# Should now print False.
print(my_list.is_empty())

# Current size is 2.
print(my_list.length)

# Remove the node containing 1.
my_list.remove(1)

# Size should now be 1.
print(my_list.length)
```
