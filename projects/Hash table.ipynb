```python
class HashTable:
    # This method runs automatically whenever we create a new HashTable object.
    def __init__(self):
        # 'collection' is the main dictionary that stores our hash table.
        # Format:
        # {
        #     hash_value: {
        #         original_key: value
        #     }
        # }
        self.collection = {}

    # ---------------- HASH FUNCTION ----------------
    def hash(self, string):
        # Variable to store the running total of Unicode values.
        total = 0

        # Go through every character in the string.
        for char in string:
            # ord(char) returns the Unicode (ASCII) value of the character.
            total += ord(char)

        # Return the final hash value.
        return total

    # ---------------- ADD ----------------
    def add(self, key, value):

        # Compute the hash value of the key.
        hashed_key = self.hash(key)

        # If this hash value does not exist yet,
        # create a new empty dictionary for it.
        if hashed_key not in self.collection:
            self.collection[hashed_key] = {}

        # Store the key-value pair inside the nested dictionary.
        #
        # Example:
        # self.collection = {
        #     424: {
        #         "golf": "sport"
        #     }
        # }
        #
        self.collection[hashed_key][key] = value

    # ---------------- REMOVE ----------------
    def remove(self, key):

        # Compute the hash of the key.
        hashed_key = self.hash(key)

        # First check if this hash bucket exists.
        if hashed_key in self.collection:

            # Then check if the actual key exists
            # inside that bucket.
            if key in self.collection[hashed_key]:

                # Delete only that key-value pair.
                del self.collection[hashed_key][key]

                # If the nested dictionary becomes empty,
                # remove the entire bucket.
                #
                # Example:
                # {
                #     424: {}
                # }
                #
                # becomes
                #
                # {}
                #
                if not self.collection[hashed_key]:
                    del self.collection[hashed_key]

    # ---------------- LOOKUP ----------------
    def lookup(self, key):

        # Compute the hash value.
        hashed_key = self.hash(key)

        # Check if the hash bucket exists.
        if hashed_key in self.collection:

            # Check if the key exists inside the bucket.
            if key in self.collection[hashed_key]:

                # Return the associated value.
                return self.collection[hashed_key][key]

        # If the key wasn't found,
        # return None.
        return None
```
